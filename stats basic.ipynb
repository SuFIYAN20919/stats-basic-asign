```python
# 2. Write a Python program to compute the variance and standard deviation of a dataset
import statistics

# Sample dataset
data = [10, 12, 23, 23, 16, 23, 21, 16]

# Manual calculation of mean
mean = sum(data) / len(data)

# Manual calculation of variance
variance = sum((x - mean) ** 2 for x in data) / len(data)  # Population variance

# Manual calculation of standard deviation
std_deviation = variance ** 0.5

print("Manual Calculation:")
print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_deviation)

# Using built-in functions (for population variance and stdev)
print("\nUsing statistics module:")
print("Variance:", statistics.pvariance(data))         # Population variance
print("Standard Deviation:", statistics.pstdev(data))  # Population standard deviation
```

```python
#  4.Implement sampling techniques like random sampling and stratified sampling
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Create a sample dataset
data = {
    'ID': range(1, 21),
    'Age': np.random.randint(18, 60, size=20),
    'Gender': np.random.choice(['Male', 'Female'], size=20),
    'Income': np.random.randint(20000, 80000, size=20)
}

df = pd.DataFrame(data)
print("Full Dataset:")
print(df)
```

```python
# Randomly sample 5 records
random_sample = df.sample(n=5, random_state=42)
print("\nRandom Sample:")
print(random_sample)
```

```python
# Stratified sampling (e.g., 30% from each gender)
stratified_sample, _ = train_test_split(df, 
                                        test_size=0.7, 
                                        stratify=df['Gender'], 
                                        random_state=42)
print("\nStratified Sample by Gender:")
print(stratified_sample)
```

```python
# 5.Write a Python function to calculate the range of a dataset
def calculate_range(data):
    """
    Calculate the range of a dataset.
    
    Parameters:
    data (list or tuple): A list or tuple of numerical values.

    Returns:
    float: The range of the dataset (max - min).
    """
    if not data:
        return None  # or raise an error
    
    return max(data) - min(data)

# Example usage
dataset = [12, 45, 23, 67, 34, 89, 10]
print("Range of dataset:", calculate_range(dataset))
```

```python
# 6.Create a dataset and plot its histogram to visualize skewness
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
# Create a positively skewed dataset using exponential distribution
data = np.random.exponential(scale=2, size=1000)

# Calculate skewness value
data_skewness = skew(data)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title(f"Histogram of Skewed Data (Skewness = {data_skewness:.2f})")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 7.Calculate skewness and kurtosis of a dataset using Python libraries
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

# Example dataset
data = [12, 15, 12, 18, 19, 22, 24, 20, 30, 45]

# Convert to Pandas Series (optional)
series = pd.Series(data)

# Calculate skewness
skewness = skew(series)
print("Skewness:", skewness)

# Calculate kurtosis
kurt = kurtosis(series)  # By default, this is 'excess kurtosis' (normal = 0)
print("Kurtosis:", kurt)
```

```python
# 8.Generate a dataset and demonstrate positive and negative skewness
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Generate datasets
np.random.seed(0)

# Positively skewed data (right skew)
positive_skew = np.random.exponential(scale=2, size=1000)

# Negatively skewed data (left skew)
negative_skew = -np.random.exponential(scale=2, size=1000)

# Calculate skewness
pos_skewness = skew(positive_skew)
neg_skewness = skew(negative_skew)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Positive skew
axes[0].hist(positive_skew, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title(f'Positively Skewed (Skewness = {pos_skewness:.2f})')

# Negative skew
axes[1].hist(negative_skew, bins=30, color='salmon', edgecolor='black')
axes[1].set_title(f'Negatively Skewed (Skewness = {neg_skewness:.2f})')

plt.tight_layout()
plt.show()
```

```python
# 9.Write a Python script to calculate covariance between two datasets
def mean(data):
    return sum(data) / len(data)

def covariance(x, y):
    if len(x) != len(y):
        raise ValueError("Datasets must be of equal length.")
    
    mean_x = mean(x)
    mean_y = mean(y)

    cov = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(len(x))) / len(x)
    return cov

# Example datasets
X = [2, 4, 6, 8, 10]
Y = [1, 3, 5, 7, 9]

# Calculate covariance
result = covariance(X, Y)
print("Covariance:", result)
```

```python
#10. Write a Python script to calculate the correlation coefficient between two datasets
# Manual and NumPy-based calculation of correlation coefficient

import numpy as np

# Sample datasets
x = [43, 21, 25, 42, 57, 59]
y = [99, 65, 79, 75, 87, 81]

# Function to compute correlation coefficient manually
def correlation_coefficient(x, y):
    n = len(x)
    if n != len(y):
        raise ValueError("Datasets x and y must be of the same length.")
    
    mean_x = sum(x) / n
    mean_y = sum(y) / n

    numerator = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n))
    denominator_x = sum((x[i] - mean_x) ** 2 for i in range(n))
    denominator_y = sum((y[i] - mean_y) ** 2 for i in range(n))

    r = numerator / (denominator_x * denominator_y) ** 0.5
    return r

# Manual calculation
r_manual = correlation_coefficient(x, y)
print(f"Correlation coefficient (manual): {r_manual:.4f}")

# Using NumPy
r_numpy = np.corrcoef(x, y)[0, 1]
print(f"Correlation coefficient (NumPy):  {r_numpy:.4f}")
```

```python
# 12. Implement and compare simple random sampling and systematic sampling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating a sample dataset of 1000 entries (e.g., population ages)
np.random.seed(42)
data = pd.DataFrame({'Age': np.random.randint(18, 80, size=1000)})

def simple_random_sample(data, sample_size):
    return data.sample(n=sample_size, random_state=42)

def systematic_sample(data, sample_size):
    k = len(data) // sample_size
    start = np.random.randint(0, k)
    indices = np.arange(start, len(data), step=k)[:sample_size]
    return data.iloc[indices]

# Set sample size
sample_size = 100

# Get samples
srs_sample = simple_random_sample(data, sample_size)
sys_sample = systematic_sample(data, sample_size)

# Compare means
print("Population Mean:", data['Age'].mean())
print("SRS Sample Mean:", srs_sample['Age'].mean())
print("Systematic Sample Mean:", sys_sample['Age'].mean())

# Plot comparison
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.hist(data['Age'], bins=20, color='gray', alpha=0.7)
plt.title('Original Data')

plt.subplot(1, 3, 2)
plt.hist(srs_sample['Age'], bins=20, color='blue', alpha=0.7)
plt.title('Simple Random Sample')

plt.subplot(1, 3, 3)
plt.hist(sys_sample['Age'], bins=20, color='green', alpha=0.7)
plt.title('Systematic Sample')

plt.tight_layout()
plt.show()
```
